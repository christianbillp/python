### Feature reduction
Since you must reduce the amount of sensors, you need to find out which ones you can get rid of. 

**Q#3** Why would PCA be useful for this?

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Do your PCA here.
data['X_std'] = StandardScaler().fit_transform(data['numerical'].values)
data['covariance'] = np.cov(data['X_std'].T)
data['eigenvalues'], data['eigenvectors'] = np.linalg.eig(data['covariance'])
data['eigenvalues_variance_explained'] = pd.Series([(i/data['eigenvalues'].sum())*100 for i in sorted(data['eigenvalues'], reverse=True)])
```

```python
data['eigenvalues_variance_explained'].plot()
```

```python
n = 0
total = 0
for item in data['eigenvalues_variance_explained']:
    if total > 90:
        print("{}, {}".format(n, total))
        break
    else:
        total += item
        n += 1
```

### Scree plot
**Q#4** How many principal components do you need to explain 90 % of the variance?

```python
n = 100
plt.figure(figsize=(10, 5))
plt.bar(range(n), data['eigenvalues_variance_explained'], alpha=0.3333, align='center', label='individual explained variance', color = 'g')
plt.axhline(y=95, linewidth=1, color='r', linestyle='dashed', label="95%")
plt.axhline(y=90, linewidth=1, color='r', linestyle='dashed', label="90%")
plt.step(range(n), np.cumsum(data['eigenvalues_variance_explained']), where='mid',label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.title('Individual and cumulative variance explained')
plt.show()
```
